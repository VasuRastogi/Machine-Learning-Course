```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Downloads\FCS\Machine Learning-A-Z-Codes-Datasets\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 4 - Simple Linear Regression\Python\Salary_Data.csv")
```

```python
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.6, random_state= 1)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

# Visualizing training data

```python
plt.scatter(X_train, y_train, color = 'blue', marker='+')
plt.plot(X_train, reg.predict(X_train), color = 'red')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Visualizing test data

```python
plt.scatter(X_test, y_test, color = 'green', marker='+')
plt.plot(X_test, y_pred, color='blue')
plt.title("Testing Data")
plt.xlabel("Year of Test Experience")
plt.ylabel("Test Salary")
```
